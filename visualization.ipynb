```python
import numpy as np 
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
## Functional method

plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
## Creating Subplot 

plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OOP Method = Create Figure objects 

## Creating figure 
fig = plt.figure()

## ([left of axis, bottom of axis, width of axis, height of axis])
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes:
    current_ax.plot (x,y)
```

```python
## Figure Size and DPI 

fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x **2, label='X Squared')
ax.plot(x, x **3, label = 'X Cubed')

## adding legend 
ax.legend(loc=0)
```

```python
## Customizing Plots
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# Changining colors 
# Changing width 
# Markers 
ax.plot(x,y,color='blue',lw=3,ls='-',marker='o')



```

```python
## Adjusting Ranges 

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='blue',lw=3,ls='-',marker='o')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

## Seaborn

```python
# Distribution Plots 

import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#Dist plot 

sns.distplot((tips['total_bill']))
```

```python
# Joint Plot = combine two different dist plots 

sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
# pairplot = compare all total numerical values 

sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
# rug plot

sns.rugplot(tips['total_bill'])
```

```python
# Categorical Plots 

import seaborn as sns
%matplotlib inline 
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
# Barplots (Averages)

sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
# Count Plot 

sns.countplot(x='sex',data=tips)
```

```python
# Box plot

sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
# Violin Plot 
# Hue arguements = legend 
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
# Strip Plot 
# Jitter to see plots 
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
# Swarm plot
# Combination of strip plot and violin plot 
# Drawback = they don't scale well with large numbers 

sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
# Matrix Plots 

import seaborn as sns
%matplotlib inline 
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
# Heat map 
# Annot arguement to add numeric values
sns.heatmap(tc,annot=True)
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='white',linewidths=1)
```

```python
# Cluster map 
# Heiarchal data 
sns.clustermap(fp,cmap='coolwarm',linecolor='white',linewidths='1')
```
